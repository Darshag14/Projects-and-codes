```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
crime = pd.read_csv("C:/Users/darsh/Downloads/crime_data.csv")
```

```python
crime.head()
```

```python
crime.shape
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_crime = norm_func(crime.iloc[:,1:])
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_crime, method='single'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'single')
```

```python
#save clusters for chart
y_hc = hc.fit_predict(df_crime)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters
```

```python
pd.concat([crime,Clusters],axis=1)
```

```python
from matplotlib import pyplot as plt
import seaborn as sn
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
crime = pd.read_csv("C:/Users/darsh/Downloads/crime_data.csv")
```

```python
crime.head()
```

```python
crime.shape
```

```python
# Normalization function 
scaler = StandardScaler()
scaled_crime = scaler.fit_transform(crime.iloc[:,1:])
```

```python
scaled_crime
```

```python
# How to find optimum number of  cluster
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
crime_clusters_new = KMeans(4, random_state=42)
crime_clusters_new.fit(scaled_crime)
```

```python
crime_clusters_new.labels_
```

```python
#Assign clusters to the data set
crime['clusterid_new'] = crime_clusters_new.labels_
```

```python
crime.groupby('clusterid_new').agg(['mean']).reset_index()
```
