```python
book_table=book1.pivot_table(index='User_ID',columns='Book_Title',values='Book_Rating').reset_index(drop=True)
```

```python
book_table
```

```python
book_table.index=book1.User_ID.unique()
```

```python
book_table.fillna(0,inplace=True)
```

```python
book_table
```

```python
book_user_similarity=1-pairwise_distances(book_table.values,metric='cosine')
```

```python
book_user_similarity
```

```python
#store results in dataframe
book_user_similarity_df=pd.DataFrame(book_user_similarity)
```

```python
book_user_similarity_df.index=book1.User_ID.unique()
book_user_similarity_df.columns=book1.User_ID.unique()
```
